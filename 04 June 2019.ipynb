```python
#Graph to plot distance and time
distance = list(range(2,11,2))
time = list(range(1,6))
plt.title('Distance Vs Time')
plt.xlabel('Time')
plt.ylabel('Dstance')
plt.grid()
plt.plot(time,distance,'go')
plt.show()
```

```python
#graph for Nmbers vs squares
n = list(range(10))
sgn = [i ** 3 for i in n]
plt.plot(n,sgn)
```

```python
#Voltage vs Time
import numpy as np
t = [i/10 for i in range(1,21)]
v = [1 + np.sin(2*np.pi*i) for i in t]
plt.grid()
plt.title("Time Vs Voltage Graph")
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.plot(t,v)
plt.savefig('images/voltage_time.png')
plt.show()
```

```python
#plot Income of Arkanses from 2005 to 2013
import pandas as pd
df = pd.read_csv('Datafiles/income.csv')
years = df.columns[2:]
incomeArkansas = df.values[3,2:]
plt.grid()
plt.title('Income of Arkanses from 2005 to 2013')
plt.xlabel("Years")
plt.ylabel("Income")
plt.plot(years,incomeArkansas)
```

```python
#Income of all states from 2005 to 2013
df
incomeallstates = []

for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
plt.title('All States Income')
plt.xlabel('year')
plt.ylabel('Income')
plt.grid()        
```
