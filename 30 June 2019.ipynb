```python
filepath = 'Datafiles/income.csv'
import pandas as pd
def readCSVdata(filepath):
    return pd.read_csv(filepath)
readCSVdata(filepath)
```

```python
incomedf = readCSVdata(filepath)
#Function to print all cums in a single line
#o/p: GEOID State 2005 2006 2007 2008 2009 2010 2011 2012 2013
def printDataFrameColumns(df):
    columns  = df.columns
    for c in columns:
        print(c,end = ' ')
    return
printDataFrameColumns(incomedf)        
```

```python
#Function to access a row based on a unique column value
def accessDataframerow(df,key):
    for row in df.values:
        if key in row:
            for item in row:
                print(item, end=' ')
    return
accessDataframerow(incomedf, 'California')    
```

```python
#Accessing a unique value based on row ,col
#income of a state in a given year
def getRowIndex(df,rowkey):
    for i in range(len(df.values)):
        if df.values[i][0] == rowkey or df.values[i][1]==rowkey:
            rowindex = i
    return rowindex

def getColumnIndex(df,colkey):
     for i in range(len(df.columns)):
        if df.columns[i] == colkey:
            columnindex = i 
def valueFromRowCol(df,rowkey,colkey):
    rowindex = getRowIndex(df,rowkey)
    columnindex = getColumnIndex(df,colkey)
    return df.values[rowindex][columnindex]

valueFromRowCol(incomedf,'California','2007')
```

```python
#Function to update the data row and column key
def updateDataFromRowColumn(df,rowkey,colkey,newval):
    rowindex = getRowIndex(df,rowkey)
    columnindex = getColumnIndex(df,colkey)
    row = df.values[rowindex]
    row[columnindex] = newval
    
    df.loc[rowindex] = row
    
    return 

updateDataFromRowColumn(incomedf,'California','2007','5700')
    
    
    
    
    
    
```
